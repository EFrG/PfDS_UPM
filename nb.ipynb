## generate portfolio allocations

```python
import itertools

portfolio_allocations = itertools.combinations_with_replacement(['ST', 'CB', 'PB', 'GO', 'CA'], 5)

#convert into list
p_all = list(portfolio_allocations)

#confirm that there are 126 combinations with repetition
print(f"There are {len(p_all)} portfolio allocation instances")
print("\n")

#now convert it into the agreed model, ST/CB/PB/GO/CA
formatted_portfolio = []
formatted_portfolio_csv = []

for el in p_all:
    #determine amount of each asset
    nST = str((el.count("ST")) * 20)
    nCB = str((el.count("CB")) * 20)
    nPB = str((el.count("PB")) * 20)
    nGO = str((el.count("GO")) * 20)
    nCA = str((el.count("CA")) * 20)

    #format string
    res = nST+"/"+nCB+"/"+nPB+"/"+nGO+"/"+nCA
    #format with "/"
    formatted_portfolio.append(res)
    #format to save csv file
    row = [nST, nCB, nPB, nGO, nCA]
    formatted_portfolio_csv.append(row)

print("ST/CB/PB/GO/CA")
for el in formatted_portfolio_csv:
    print(el)
```

## Save to csv portfolio_allocations.csv

```python
import csv

#specific for windows, linux should end lines without \n
with open('portfolios/portfolio_allocations.csv', 'w', newline='') as f:
    #define writer
    writer = csv.writer(f)
    #header
    header = ["ST", "CB", "PB", "GO", "CA"]
    writer.writerow(header)

    for el in formatted_portfolio_csv:
        writer.writerow(el)
```
